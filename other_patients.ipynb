```python
import pandas as pd
import numpy as np
import pickle
import os

root = "/home/rodriguez/Documents/UPJV/Sujets_Pathologiques/Luisin_L/"
# Create a list with the names of the csv files in root folder
csv_files = [f for f in os.listdir(root) if f.endswith('.csv')]
print(csv_files)
```

```python
luisin_df = pd.DataFrame(columns=['Name', 'Content'])
for csv in csv_files:
    filepath = root + csv
    print(filepath)
    # Try to read csv file and store its content as a numpy array
    try:
        csv_content = pd.read_csv(filepath, header=None, skiprows=5, usecols=[i+2 for i in range(324)])
        csv_array = csv_content.to_numpy()
        luisin_df = pd.concat([luisin_df, pd.DataFrame({'Name': csv, 'Content': [csv_array]})], ignore_index=True)
    except:
        print("Error while reading csv file. File: " + csv + " ignored.")


luisin_df.head()
```
