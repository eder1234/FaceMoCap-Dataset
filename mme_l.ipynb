```python
import pandas as pd
import numpy as np
import pickle
import os

root = "/home/rodriguez/Documents/UPJV/Sujets_Pathologiques/Madame_L/"
# Create a list with the names of the folders in the root directory
folders = os.listdir(root)
print(folders)
```

```python
files_list = []
for folder in folders:
    # Create a list with the names of the files in the folder
    files = os.listdir(root + folder)
    for file in files:
        files_list.append(root + folder + "/" + file)
print(files_list)
```

```python
mme_l_df = pd.DataFrame(columns=['File name', 'Original SPC'])
for csv in files_list:
    filepath = csv
    print(filepath)
    # Try to read csv file and store its content as a numpy array
    try:
        csv_content = pd.read_csv(filepath, header=None, skiprows=5, usecols=[i+2 for i in range(324)])
        csv_array = csv_content.to_numpy()
        mme_l_df = pd.concat([mme_l_df, pd.DataFrame({'File name': csv, 'Original SPC': [csv_array]})], ignore_index=True)
    except:
        print("Error while reading csv file. File: " + csv + " ignored.")
mme_l_df.head()
```

```python
for i in range(len(mme_l_df)):
    filename = mme_l_df["File name"][i][70:]
    mme_l_df["File name"][i] = filename
    print(filename)
```

```python
mme_l_df.head(20)
```

```python
mme_l_df["Name"] = "LB"
mme_l_df["Pathology"] = 'Transplantation'
mme_l_df["MoCap Platform"] = "IFF"
mme_l_df["Experimentator"] = "Félix"
mme_l_df.head()
```

```python
mme_l_df["Date"] = ""
for index, row in mme_l_df.iterrows():
    date_str = files_list[index][61:69]
    if date_str == '20190418':
        mme_l_df["Date"][index] = '18-04-2019'
        mme_l_df["MoCap Platform"][index] = "Unknown"
    elif date_str == '20140710':
        mme_l_df["Date"][index] = '10-07-2014'
        mme_l_df["MoCap Platform"][index] = "UTC"
    elif date_str == '20150225':
        mme_l_df["Date"][index] = '25-02-2015'
        mme_l_df["MoCap Platform"][index] = "UTC"
    elif date_str == '20171205':
        mme_l_df["Date"][index] = '05-12-2017'
        mme_l_df["MoCap Platform"][index] = "UTC"
    elif date_str == '20211011':
        mme_l_df["Date"][index] = '01-11-2021'
        mme_l_df["MoCap Platform"][index] = "IFF"
    elif date_str == '20230208':
        mme_l_df["Date"][index] = '08-02-2023'
        mme_l_df["MoCap Platform"][index] = "IFF"
    else:
        print("Unknown date: " + date_str)
        mme_l_df["Date"][index] = 'Unknown'
mme_l_df.head()
```

```python
mme_l_df["Movement"] = ""
mme_l_df["Repetitive"] = False
mme_l_df.head()
```

```python
M_list = []
for index, row in mme_l_df.iterrows():
    try:
        M = int(mme_l_df["File name"][index][-5])
        M_list.append(M)
    except:
        M = ""
        print("Error while reading M value. File: " + mme_l_df["File name"][index] + " ignored.")
        M_list.append(M)
```

```python
mme_l_df["Movement"] = M_list
```

```python
# Set "Movement" = 5 where "File name" == 1_M5b.csv
mme_l_df["Movement"][mme_l_df["File name"] == "1_M5b.csv"] = 5
```

```python
# Set "Movement" = 3 where "File name" == 1_M5b.csv
mme_l_df["Movement"][mme_l_df["File name"] == "PatientL02_M3bis.csv"] = 3
```

```python
# Save the dataframe as a pickle file
mme_l_df.to_pickle("mme_l_df.pkl")
```
