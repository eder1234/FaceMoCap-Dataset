```python
import pickle
import pandas as pd
import numpy as np
import os
```

```python
# Create a dataframe for each csv file which 'Name' column contains the csv file name and 'Content' column contains the csv file content
pathology_df = pd.DataFrame()
root_path = "/home/rodriguez/Documents/UPJV/Sujets Pathologiques/Fichier_PF/"
# Create a variable that reads all csv files in the root_path
csv_list = os.listdir(root_path)
for csv in csv_list:
    filepath = root_path + csv
    print(filepath)
    # Try to read csv file and store its content as a numpy array
    try:
        csv_content = pd.read_csv(filepath, header=None, skiprows=5, usecols=[i+2 for i in range(324)])
        csv_array = csv_content.to_numpy()
        pathology_df = pd.concat([pathology_df, pd.DataFrame({'Name': csv, 'Content': [csv_array]})], ignore_index=True)
    except:
        print("Error while reading csv file. File: " + csv + " ignored.")


pathology_df.head()


```

```python
pathology_df["Pathology"] = True
pathology_df["MoCap platform"] = "UTC"
pathology_df["Experimentator"] = "Mattieu"
pathology_df["Movement"] = ""
# Create a boolean variable that determines if the name contains 'x' in its name
pathology_df["Repetitive"] = pathology_df["Name"].str.contains("x")  
pathology_df.head(20)
```

```python
# Create a list that contains all the names in the dataframe
pathology_names = pathology_df["Name"].tolist()
```

```python
import regex as re
# Create a regex variable that finds the number next to the letter 'M' in the name
regex = r"M(\d+)"
# Create a list that contains the movement number for each name
movement_list = []
for name in pathology_names:
    try:
        movement_list.append(re.search(regex, name).group(0))
    except:
        movement_list.append("")

print(movement_list)
 
```

```python
# Add the movement list to the dataframe
pathology_df["Movement"] = movement_list
pathology_df.head(20)
```

```python
# Save the dataframe as a pickle file
pathology_df.to_pickle("pathology_df.pkl")
```
