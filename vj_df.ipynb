```python
import os

# Create a list that contains the folders in /home/rodriguez/Documents/UPJV/Sujets_Pathologiques/Varin_J/
root = '/home/rodriguez/Documents/UPJV/Sujets_Pathologiques/Varin_J/'
folders = os.listdir(root)
print(folders)
```

```python
files_list = []
for folder in folders:
    # Create a list with the names of the files in the folder
    files = os.listdir(root + folder)
    for file in files:
        files_list.append(root + folder + "/" + file)
print(files_list)
```

```python
import pandas as pd

vj_df = pd.DataFrame(columns=['File name', 'Original SPC'])
for csv in files_list:
    filepath = csv
    print(filepath)
    # Try to read csv file and store its content as a numpy array
    try:
        csv_content = pd.read_csv(filepath, header=None, skiprows=5, usecols=[i+2 for i in range(324)])
        csv_array = csv_content.to_numpy()
        vj_df = pd.concat([vj_df, pd.DataFrame({'File name': csv, 'Original SPC': [csv_array]})], ignore_index=True)
    except:
        print("Error while reading csv file. File: " + csv + " ignored.")
vj_df.head()
```

```python
for i in range(len(vj_df)):
    filename = vj_df["File name"][i][69:]
    vj_df["File name"][i] = filename
    print(filename)
```

```python
vj_df["Name"] = "VJ"
vj_df["Pathology"] = 'Facial palsy'
vj_df["MoCap Platform"] = "IFF"
vj_df["Experimentator"] = "Félix"
vj_df.head()
```

```python
vj_df["Date"] = ""
for index, row in vj_df.iterrows():
    date_str = files_list[index][60:68]
    if date_str == '20220315':
        vj_df["Date"][index] = '03-15-2022'
    elif date_str == '20230124':
        vj_df["Date"][index] = '24-01-2023'
    elif date_str == '20220913':
        vj_df["Date"][index] = '13-09-2022'
    else:
        print("Unknown date: " + date_str)
        vj_df["Date"][index] = 'Unknown'
    
vj_df.head()
```

```python
vj_df["Repetitive"] = False
```

```python
M_list = []
for index, row in vj_df.iterrows():
    try:
        M = int(vj_df["File name"][index][-8])
        M_list.append(M)
    except:
        M = ""
        print("Error while reading M value. File: " + vj_df["File name"][index] + " ignored.")
        M_list.append(M)
vj_df["M"] = M_list
```
