```python
import os

# Create a list that contains the folders in '/home/rodriguez/Documents/Sujets Sains'
root = '/home/rodriguez/Documents/Sujets Sains'
frs_folders = os.listdir(root)
print(frs_folders)
```

```python
# Remove 'SIMOVI_Sain' from the list
frs_folders.remove('SIMOVI_Sain')
frs_folders.remove("Readme.md")
```

```python
import pandas as pd

frs_df = pd.DataFrame()
frs_error_list = []
for folder in frs_folders:
    filepath = os.path.join(root, folder)
    csv_files = os.listdir(filepath)
    for csv in csv_files:
        csv_path = os.path.join(filepath, csv)
        try:
            csv_content = pd.read_csv(csv_path, header=None, skiprows=5, usecols=[i+2 for i in range(324)])
            
            csv_array = csv_content.to_numpy()
            # If "x" is in csv, set repetitive = True
            if "x" in csv:
                repetitive = True
            else:
                repetitive = False
            frs_df = pd.concat([frs_df, pd.DataFrame({'File name': csv, 'Original SPC': [csv_array], 
                                                      "Name": folder, "Experimentator": "François-Régis", 
                                                      'MoCap platform': "UTC", "Pathology": False,
                                                      "Repetitive": repetitive})], 
                                                      ignore_index=True)
        except:
            print("Error while reading csv file. File: " + csv_path + " ignored.")
            frs_error_list.append(csv_path)
frs_df["Movement"] = ""
frs_df["Date"] = ""
```

```python
frs_df.head(20)
```

```python
# Determine the movement and the date of each file
for index, row in frs_df.iterrows():
    M0X = row["File name"][-7:-4]
    MX = row["File name"][-6:-4]
    MXbis = row["File name"][-9:-4]
    MX_ = row["File name"][-8:-5]
    MXx = row["File name"][-8:-5]
    
    if (M0X == 'M05' or MX == 'M5' or MXbis == 'M5bis' or MX_ == 'M5_' or MXx == 'M5x'):
        frs_df.loc[index, "Movement"] = 5
    if (M0X == 'M04' or MX == 'M4' or MXbis == 'M4bis' or MX_ == 'M4_' or MXx == 'M4x'):
        frs_df.loc[index, "Movement"] = 4
    if (M0X == 'M03' or MX == 'M3' or MXbis == 'M3bis' or MX_ == 'M3_' or MXx == 'M3x'):
        frs_df.loc[index, "Movement"] = 3
    if (M0X == 'M02' or MX == 'M2' or MXbis == 'M2bis' or MX_ == 'M2_' or MXx == 'M2x'):
        frs_df.loc[index, "Movement"] = 2
    if (M0X == 'M01' or MX == 'M1' or MXbis == 'M1bis' or MX_ == 'M1_' or MXx == 'M1x'):
        frs_df.loc[index, "Movement"] = 1
    

```

```python
import os
import pandas as pd

# Create felix_df
f_path = "/home/rodriguez/Documents/Sujets Sains/SIMOVI_Sain/AF"
f_folders = os.listdir(f_path)
f_folders.remove("Suivi labelisation.xlsx")
print(f_folders)
```

```python
felix_df = pd.DataFrame()
felix_error_list = []
for csv in f_folders:
    csv_path = os.path.join(f_path+'/', csv)
    folder = "AF"
    try:
        csv_content = pd.read_csv(csv_path, header=None, skiprows=5, usecols=[i+2 for i in range(324)])
        
        csv_array = csv_content.to_numpy()
        # If "x" is in csv, set repetitive = True
        if "x" in csv:
            repetitive = True
        else:
            repetitive = False
        felix_df = pd.concat([felix_df, pd.DataFrame({'File name': csv, 'Original SPC': [csv_array], 
                                                    "Name": folder, "Experimentator": "Félix", 
                                                    'MoCap platform': "IFF", "Pathology": False,
                                                    "Repetitive": repetitive})], 
                                                    ignore_index=True)
    except:
        print("Error while reading csv file. File: " + csv_path + " ignored.")
        felix_error_list.append(csv_path)
felix_df["Movement"] = ""
felix_df["Date"] = ""
felix_df.head(20)
```

```python
for index, row in felix_df.iterrows():
    M = int(row["File name"][1])
    felix_df.loc[index, "Movement"] = M
felix_df.head(19)
```

```python
felix_df.to_pickle("felix_df.pkl")
```

```python
# Load frs_df
frs_df = pd.read_pickle("frs_df.pkl")
frs_df.head(20)
```

```python
dates_dict = {'29-11-2013': ['CG01'],
              '13-05-2014': ['CN01', 'CG02', 'JB01', 'AC01', 'FRS01'],
              '11-06-2014': ['CN02', 'FRS02', 'CG03', 'AA01', 'NL01', 'CR01', 'RL01', 'BT01', 'LN01'],
              '12-05-2015': ['CD01', 'TP01', 'SD01', 'MP01'],
              '29-06-2015': ['FRS03', 'CG04'],
              '13-07-2015': ['FRS04', 'CG05'],
              '15-10-2015': ['CG06', 'RS01', 'MM01', 'AG01', 'CAD01']}
```

```python
# In frs_df, set the date (key) according to the name (value) in dates_dict
for key, value in dates_dict.items():
    for name in value:
        frs_df.loc[frs_df["Name"] == name, "Date"] = key

frs_df.head(20)
```
