```python
import pickle
import pandas as pd

# Load luisin_df, mme_l_df, vj_df, and patholoy_df
with open('luisin_df.pkl', 'rb') as f:
    luisin_df = pickle.load(f)
with open('mme_l_df.pkl', 'rb') as f:
    mme_l_df = pickle.load(f)
with open('vj_df.pkl', 'rb') as f:
    vj_df = pickle.load(f)
with open('pathology_df.pkl', 'rb') as f:
    pathology_df = pickle.load(f)
```

```python
luisin_df.rename(columns={'Aquisition date': 'Date'}, inplace=True) 
luisin_df.head()
```

```python
# Rename column "MoCap Platform" to "MoCap platform"
mme_l_df.rename(columns={'MoCap Platform': 'MoCap platform'}, inplace=True) 
mme_l_df.head()
```

```python
vj_df.rename(columns={'MoCap Platform': 'MoCap platform'}, inplace=True) 
vj_df.head()
```

```python
pathology_df.head()
```

```python
# Concatenate luisin_df, mme_l_df, pathology_df, and vj_df into a new dataframes named patients_df
patients_df = pd.concat([luisin_df, mme_l_df, pathology_df, vj_df], axis=0)
patients_df.head(50)
```

```python
# Save patients_df as a pickle file
with open('patients_df.pkl', 'wb') as f:
    pickle.dump(patients_df, f)
```
